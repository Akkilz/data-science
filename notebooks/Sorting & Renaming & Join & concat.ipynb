```python
import pandas as pd

import statsmodels.api as sm  #(To access mtcars dataset)
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data

mtcars.head()
```

```python


# Sort a data frame by a column with df.sort_values()

mtcars.sort_values(by = ["mpg"],   # Row or columns names to sort by
                   axis = 0,       # Sort Rows axis = 0
                   ascending=True) # Sort ascending or descending?

```

```python
# Sort by multiple columns

mtcars.sort_values(by = ["cyl", "mpg"],     # Multiple column sorting
                   axis = 0,                # Sort rows; axis = 0
                   ascending=[True, True]) # Sort ascending or descending?
```

```python
import pandas as pd

data = pd.DataFrame({"char": ["Goku","Vegeta", "Nappa","Gohan","Piccolo"], 
                     "pl": [12000, 16000, 4000, 1500, 3000],
                     "ucl": ["orange", "blue", "black", "orange", "purple"]})

data
```

```python
# Rename columns with df.rename()

data.rename(columns={'char':'character',
                     'pl': 'power level',
                     'ucl': 'uniform color'}, inplace = True)

data
```

```python
import pandas as pd

data1 = pd.DataFrame({"character": ["Goku","Vegeta", "Nappa","Gohan","Piccolo"], 
                      "power level": [12000, 16000, 4000, 1500, 3000],
                      "uniform color": ["orange", "blue", "black", "orange", "purple"]})

data2 = pd.DataFrame({"character": ["Gohan","Goku", "Tien","Krillin","Yamcha"], 
                      "power level": [1500, 12000, 2000, 2000, 1500],
                      "unif_color": ["purple", "orange", "green", "orange", "orange"]})

data1
```

```python
data2
```

```python
# Use df.merge() to join data frames

data1.merge(data2,          
            how = "inner",  # join type
            on="character") # key column to join on
```

```python
# Use df.merge() to join data frames

data1.merge(data2,          
            how = "left",  # join type
            on="character") # key column to join on
```

```python
# Use df.merge() to join data frames

data1.merge(data2,          
            how = "outer",  # join type
            on="character") # key column to join on
```

```python
# Join on multiple key columns
data1.merge(data2,          
            how = "outer",                                     # join type
            on = ["character", "power level"])     # join on multiple cols
```

```python

# Rename columns so that they match
data2.rename(columns={'unif_color':'uniform color'}, inplace=True)

# Then join the data
data1.merge(data2,          
            how = "outer",                            
            on = ["character", "power level", "uniform color"])  
```

```python
# Join columns with different names

data1.merge(data2,          
            how = "outer",                            
            on_right = ["character", "power level", "uniform color"],
            on_left = ["character", "power level", "unif_color"]) 
```

```python
import pandas as pd

data = pd.DataFrame({"character": ["Goku","Vegeta", "Nappa","Gohan","Piccolo"], 
                     "power level": [12000, 16000, 4000, 1500, 3000]})

data
```

```python
new_rows = pd.DataFrame({"character": ["Tien","Yamcha", "Krillin"], 
                     "power level": [2000, 1600, 2000]})

new_rows
```

```python
# Concatenate Data Frames by Rows

data2 = pd.concat([data, new_rows], # List of data Frames to concatenate
          axis=0,                   # Axis = 0 to concat by row
          ignore_index=True)         

data2
```

```python
# Concatenate Data Frames by Rows

data2 = pd.concat([data, new_rows], # List of data Frames to concatenate
          axis=0)                 # Axis = 0 to concat by row
                   

data2
```

```python

new_cols = pd.DataFrame({"uniform color": ["orange", "blue", "black", "orange", 
                                           "purple", "green", "orange", "orange"],
                         "species":["saiyan","saiyan","saiyan","half saiyan",
                                    "namak","human","human","human"]})

new_cols
```
